```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("googleplaystore.csv", nrows=1000)
```

```python
df
```

```python
df.shape
```

```python
x = df.Rating
y = df.Reviews

plt.scatter(x, y)
```

```python
plt.figure(figsize=(16, 9))
plt.scatter(x, y, color="r", marker="*", s=100, alpha=0.5, linewidth=10, edgecolor="g")
plt.title("Google Play Store apps scatter plot")
plt.xlabel("ratings")
plt.ylabel("reviews & installs")
```

```python
#multiple scatter plot

plt.figure(figsize=(16, 9))
plt.scatter(x, y, color="r", marker="*", s=100, alpha=0.5, linewidth=10, edgecolor="g")

plt.scatter(x, df.Installs, color="r", marker="o", s=100, alpha=0.5, linewidth=10, edgecolor="c")

plt.title("Google Play Store apps scatter plot")
plt.xlabel("ratings")
plt.ylabel("reviews")
```
