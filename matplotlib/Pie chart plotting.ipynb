```python
 
    classes = ["Python", "R", 'Machine learning', 'Artificial Intelligence', 'Data Science']
class1_students = [45, 15, 35, 25, 30]
colors = ['c', 'b', 'r', 'y', 'g']
textprops = {"fontsize":15}  #it is always in dic form


explode = [0.03, 0, 0.1, 0, 0]
wedgeprops = {'linewidth':1.2, 'edgecolor':'k'}
plt.pie(class1_students, labels = classes, explode=explode, colors = colors, autopct="%0.2f%%", shadow=True, radius=1.5, startangle = 90,
       textprops = textprops, labeldistance=1.2,wedgeprops = wedgeprops, rotatelabels=True )


plt.title("percentage of students in each class")
plt.legend()
plt.show()
```
