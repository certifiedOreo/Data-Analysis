#  Line Plot


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = np.arange(1, 31)
temperature = np.random.uniform(35.1, 49.9, 30)
temperature
```

```python
plt.plot(days, temperature)
plt.title("Delhi Temperature")
plt.xlabel("Days")
plt.ylabel("temperature")
```

```python
#manually change the axis starting and ending value
days = np.arange(1, 31)
temperature = np.random.uniform(35.1, 49.9, 30)
plt.plot(days, temperature)
plt.axis([0, 30, 0, 50])
plt.title("Delhi Temperature")
plt.xlabel("Days")
plt.ylabel("temperature")
```

```python
#coloring and formatting the line graph
days = np.arange(1, 31)
temperature = np.random.uniform(35.1, 49.9, 30)
plt.plot(days, temperature, color="g", marker="o", linestyle="--", linewidth=2,
        markersize = 10)
plt.axis([0, 30, 0, 50])
plt.title("Delhi Temperature")
plt.xlabel("Days")
plt.ylabel("temperature")
```

```python
#shortcut to draw the above same graph
from matplotlib import style
days = np.arange(1, 31)
temperature = np.random.uniform(35.1, 49.9, 30)
style.use("ggplot")
plt.plot(days, temperature, "go--", linewidth=2,
        markersize = 10)
plt.grid(color="k", linestyle="-", linewidth=2)
plt.axis([0, 30, 0, 55])
plt.title("Delhi Temperature", fontsize =15)
plt.xlabel("Days", fontsize =13)
plt.ylabel("temperature", fontsize =13)
plt.legend(["Temp line"], loc=4)
```

```python
days = np.arange(1, 31)
delhi_temperature = np.random.uniform(35.1, 49.9, 30)
mumbai_temperature = np.random.uniform(35.1, 49.9, 30)
style.use("ggplot")
plt.plot(days, delhi_temperature, "go--", linewidth=2,
        markersize = 10, label="delhi temp")

plt.plot(days, mumbai_temperature, "mo:", linewidth=2,
        markersize = 10, label="mumbai temp")

plt.grid(color="k", linestyle="-", linewidth=2)
plt.axis([0, 30, 0, 55])
plt.title("Delhi & Mumbai Temperature", fontsize =15)
plt.xlabel("Days", fontsize =13)
plt.ylabel("temperature", fontsize =13)
plt.legend(loc=4)
```
