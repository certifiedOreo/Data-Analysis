```python
style.use("ggplot")
plt.figure(figsize = (16, 9))
plt.bar(classes, class1_students, width=0.5, align = "edge", color='k', edgecolor="m"
       , linewidth = 4, alpha = 0.9, linestyle = "--", label = "class 1 students", visible = True)
plt.title("Bar Chart of IAIP class", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No. of students", fontsize=15)
```

```python
style.use("ggplot")
plt.figure(figsize = (16, 9))
plt.bar(classes, class1_students, width=0.5, align = "edge", color='k', edgecolor="m"
       , linewidth = 4, alpha = 0.9, linestyle = "--", label = "class 1 students", visible = True)


plt.bar(classes, class2_students, width=0.5, align = "edge", color='y', edgecolor="g"
       , linewidth = 4, alpha = 0.9, linestyle = "--", label = "class 2 students", visible = True)



plt.title("Bar Chart of IAIP class", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No. of students", fontsize=15)
```

```python
#seperate and non-overlapping bar graph of each class
style.use("ggplot")
plt.figure(figsize = (16, 9))
classes_index = np.arange(len(classes))
width = 0.2
plt.bar(classes_index, class1_students, width, align = "edge", color='b'
       , linewidth = 4, alpha = 0.9, label = "class 1 students", visible = True)


plt.bar(classes_index+width, class2_students, width, align = "edge", color='y'
       , linewidth = 4, alpha = 0.9, label = "class 2 students", visible = True)


plt.bar(classes_index+width+width, class3_students, width, align = "edge", color='g'
       , linewidth = 4, alpha = 0.9, label = "class 3 students", visible = True)

plt.xticks(classes_index+width, classes, rotation=20)
plt.title("Bar Chart of IAIP class", fontsize=18)
plt.xlabel("Classes", fontsize=15)
plt.ylabel("No. of students", fontsize=15)
plt.legend()
```
