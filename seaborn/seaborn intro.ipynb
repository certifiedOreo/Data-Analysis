```python
days = np.arange(1, 31)
temperature = np.random.uniform(35.1, 49.9, 30)
temp_df = pd.DataFrame({"days": days, "temperature":temperature})
sns.lineplot(x="days", y= "temperature", data = temp_df)
plt.show()
```

```python
tips_df = sns.load_dataset("tips")
tips_df
```

```python
tips_df.shape
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x="total_bill", y = "tip", data = tips_df, hue="sex", style="sex", palette ="hot")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x="total_bill", y = "tip", data = tips_df, hue="sex", style="sex", palette ="hot", dashes=False)
plt.show()
```

```python

plt.figure(figsize=(16, 9))
sns.lineplot(x="total_bill", y = "tip", data = tips_df, hue="sex", style="sex", palette ="hot", dashes=False, markers=["o","<"], legend="brief")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x="total_bill", y = "tip", data = tips_df, hue="sex", style="sex", palette ="hot", dashes=False, markers=["o","<"], legend="brief")
plt.xlabel("Size")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.lineplot(x="size", y = "total_bill", data = tips_df, hue="sex", style="sex", palette ="hot", dashes=False, markers=["o","<"], legend="brief")
plt.title("Line plot", fontsize=20)
plt.xlabel("Size", fontsize=15)
plt.ylabel("Total Bill", fontsize=15)
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(16, 9))
sns.lineplot(x="size", y = "total_bill", data = tips_df, hue="sex", style="sex", palette ="hot", dashes=False, markers=["o","<"], legend="brief")
plt.title("Line plot", fontsize=20)
plt.xlabel("Size", fontsize=15)
plt.ylabel("Total Bill", fontsize=15)
plt.show()
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(16, 9))
sns.lineplot(x="size", y = "total_bill", data = tips_df, hue="day", style="day", palette ="hot", dashes=False, markers=["o","<",">", "^"], legend="brief")
plt.title("Line plot", fontsize=20)
plt.xlabel("Size", fontsize=15)
plt.ylabel("Total Bill", fontsize=15)
plt.show()
```
