```python
sns.histplot(tips_df["total_bill"], kde=True, color="r")
plt.show()
```

```python
sns.histplot( tips_df["total_bill"], kde=True, color="r")
plt.show()
```

```python
plt.figure(figsize=(16, 9))
sns.histplot( tips_df["total_bill"], kde=True, color="r",stat="count")
plt.title("Bill histogram", fontsize=20)
sns.set(style="darkgrid")
plt.show()
```

```python
tips_df.total_bill.sort_values()
```

```python
#GRAPH UNDER DESIRED RANGE OF VALUES DEPENDENT VARIABLE
bins = [1, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

plt.figure(figsize=(16, 9))
sns.histplot( tips_df["total_bill"], kde=True, color="r",stat="count", bins = bins)
plt.xticks(bins)
plt.title("Bill histogram", fontsize=20)
sns.set(style="darkgrid")
plt.show()
```

```python
#hist keywords
plt.figure(figsize=(16, 9))
sns.histplot( tips_df["total_bill"], cbar = {
    'color' :'#DC143C',
    'edgecolor': '#aaff00',
    'linewidth':5
    
})
plt.title("Bill histogram", fontsize=20)
sns.set(style="darkgrid")
plt.show()
```
